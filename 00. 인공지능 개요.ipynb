# 인공지능(Artificial Intelligence)
- 특정 분야를 지칭하는 것 X, **지능적 요소**가 포함된 기술을 총칭
- 인공지능의 한 분야로 **머신러닝**이 존재
    - 데이터에서 모델을 스스로 찾아내는 기법
    - 심층 신경망을 이용한 머신러닝은 **딥러닝**


# 머신러닝( Machine Learning )
- 데이터 입력 -> **처리과정** -> 출력
- 우리는 지도, 비지도학습에 중점을 두면서 수업
- sklearn을 통해 배우게 될 것
- 입력 데이터: 특성 데이터
- 출력 데이터: 라벨<br>


### 1. 지도학습( Supervised Learning )
- 문제와 정담을 모두 알려주고 공부시키는 방법
- 특정데이터와 라벨 존재
- 예측, 분류, 회귀, 추정
- 알고리즘
    - 분류 : 이산데이터
        - KNN
        - Decision Tree
        - Logistic Regression
        - Random forest
        - Support Vector Machine
        - ANN( Arificial Neural Network )
    - 예측 : 연속형 데이터
        - Linear Regression
        - Regression Tree
        - Support Vector Regression
        - ANN

### 2. 비지도 학습( Unsupervised learning )
- 답을 가르쳐주지 않고 공부시키는 방법
- 특성데이터만 존재
- 연관 규칙, 군집, 차원 축소
    - 차원 축소
        - PCA
    - 군집화
        - K-means Clustering
        - SOM
    - 연관성 규칙 발견
        - MBA

### 3. 강화학습
- 보상을 통해 상은 최대화, 벌은 최소화 방향으로 행위를 강화하는 학습
- 보상
